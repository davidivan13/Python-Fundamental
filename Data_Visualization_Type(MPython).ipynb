#Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("tips.csv")

plt.figure(figsize=(10,5))
# Scatter plot with day against tip
plt.plot(data['tip'], marker = 'o', linestyle='-.')
# plt.plot(data['size'])

# Adding Title to the Plot
plt.title("Line Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.bar(x, y)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
```

Kustomisasi yang tersedia untuk Bar Chart :
- color: Untuk wajah bar
- edgecolor: Warna tepi bilah
- linewidth: Lebar tepi bilah
- width: Lebar bilah

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.bar(x, y, color='yellow', edgecolor='black',
		linewidth=2)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
```

# Histogram

syntax:

matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype=’bar’, align=’mid’, orientation=’vertical’, rwidth=None, log=False, color=None, label=None, stacked=False, \*, data=None, \*\*kwargs)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['total_bill']

# plotting the data
plt.hist(x)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Frequency')

# Adding label on the x-axis
plt.xlabel('Total Bill')

plt.show()
```

Kustomisasi yang tersedia untuk Histogram

- bins: Jumlah tempat sampah dengan lebar yang sama
- color: Untuk mengubah warna wajah
- edgecolor: Warna tepi
- linestyle: Untuk garis tepi
-  alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['total_bill']

# plotting the data
plt.hist(x, bins=25, color='green', edgecolor='black',
		linestyle='--', alpha=0.5)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Frequency')

# Adding label on the x-axis
plt.xlabel('Total Bill')

plt.show()
```

#Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.scatter(x, y)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
```

Grafik ini bisa lebih bermakna jika kita bisa menambahkan warna dan juga mengubah ukuran titik-titiknya. Kita dapat melakukan ini dengan menggunakan parameter c dan s masing-masing dari fungsi scatter.

Kustomisasi yang tersedia untuk scatter adalah –

- s: ukuran penanda (bisa skalar atau array dengan ukuran sama dengan ukuran x atau y)
- c: warna urutan warna untuk spidol
- marker: gaya penanda
- linewidths: lebar batas penanda
- edgecolor: warna batas penanda
- alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)


Kita juga dapat menampilkan bilah warna menggunakan metode colorbar()

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.scatter(x, y, c=data['size'],  s=data['total_bill'], marker='D', alpha=0.5)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.colorbar()

plt.show()
```

#Pie Chart

syntax : matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

# plotting the data
plt.pie(data, labels=cars)

# Adding title to the plot
plt.title("Car data")

plt.show()
```

Kustomisasi yang tersedia untuk pie chart adalah –

- explode: Memindahkan irisan plot
- autopct: Beri label baji dengan nilai numeriknya.
- color: Atribut digunakan untuk memberikan warna pada irisan.
- shadow: Digunakan untuk membuat bayangan baji.

```python
import matplotlib.pyplot as plt
import pandas as pd


# initializing the data
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 13, 35, 15, 12]

explode = [0.1, 0.5, 0, 0, 0]

colors = ( "orange", "cyan", "yellow",
		"grey", "green",)

# plotting the data
plt.pie(data, labels=cars, explode=explode, autopct='%.3f%%',
		colors=colors, shadow=True)

plt.show()
```
