## Scatterplot

# Customize Plot in Seaborn

## Mengubah estetika figur

set_style() digunakan untuk mengatur estetika plot. Artinya mempengaruhi hal-hal seperti warna sumbu, apakah grid aktif atau tidak, atau elemen estetika lainnya. Ada lima tema yang tersedia di Seaborn.

- darkgrid
- whitegrid
- dark
- white
- tick

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# changing the theme to dark
sns.set_style("dark")
plt.show()
```

## Penghapusan spine

spines adalah garis yang mencatat batas data dan menghubungkan tanda centang sumbu. Itu dapat dihapus menggunakan metode despine().

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()
plt.show()
```

# Mengubah Ukuran Figur
Ukuran gambar dapat diubah menggunakan metode figure() dari Matplotlib. figure() metode membuat gambar baru dengan ukuran tertentu yang diteruskan dalam parameter figsize.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# changing the figure size
plt.figure(figsize = (2, 4))

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()

plt.show()
```

## Scaling plot

Dilakukan dengan menggunakan metode set_context(). Ini memungkinkan kita untuk mengganti parameter default. Ini memengaruhi hal-hal seperti ukuran label, garis, dan elemen plot lainnya, tetapi bukan gaya keseluruhan. Konteks dasarnya adalah “notebook”, dan konteks lainnya adalah “paper”, “talk”, dan “poster”. font_scale mengatur ukuran font.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Setting the scale of the plot
sns.set_context("paper")

plt.show()
```

## Mengatur Gaya Sementara
metode axes_style() digunakan untuk mengatur gaya sementara. Ini digunakan bersama dengan pernyataan with.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")


def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

with sns.axes_style('darkgrid'):

	# Adding the subplot
	plt.subplot(211)
	plot()

plt.subplot(212)
plot()
```

# Color Palette

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# current colot palette
palette = sns.color_palette()

# plots the color palette as a
# horizontal array
sns.palplot(palette)

plt.show()
```

##Palet Warna Divergen
Jenis palet warna ini menggunakan dua warna yang berbeda dimana setiap warna menggambarkan titik yang berbeda mulai dari titik yang sama di kedua arah. Pertimbangkan rentang -10 hingga 10 sehingga nilai dari -10 hingga 0 mengambil satu warna dan nilai dari 0 hingga 10 mengambil warna lain.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# current colot palette
palette = sns.color_palette('PiYG', 11)

# diverging color palette
sns.palplot(palette)

plt.show()
```

Dalam contoh di atas, kita telah menggunakan palet warna divergen bawaan yang menunjukkan 11 titik warna yang berbeda. Warna di sebelah kiri menunjukkan warna pink dan warna di sebelah kanan menunjukkan warna hijau.



## Mengurutkan Palet Warna
Palet sekuensial digunakan di mana distribusi berkisar dari nilai yang lebih rendah ke nilai yang lebih tinggi. Untuk melakukan ini, tambahkan karakter 's' ke warna yang dilewatkan di palet warna.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# current colot palette
palette = sns.color_palette('Greens', 11)

# sequential color palette
sns.palplot(palette)

plt.show()
```

##Mengatur Palet Warna default
set_palette() metode digunakan untuk mengatur palet warna default untuk semua plot. Argumen untuk color_palette() dan set_palette() sama. set_palette() mengubah parameter matplotlib default.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the default color palette
sns.set_palette('vlag')
plt.subplot(211)

# plotting with the color palette
# as vlag
plot()

# setting another default color palette
sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

# Multipleplots Seaborn

## Metode 1: Menggunakan metode FacetGrid()



- Kelas FacetGrid membantu dalam memvisualisasikan distribusi satu variabel serta hubungan antara beberapa variabel secara terpisah dalam himpunan bagian dari kumpulan data Anda menggunakan beberapa panel.
- Sebuah FacetGrid dapat digambar hingga tiga dimensi ? baris, col, dan hue. Dua yang pertama memiliki korespondensi yang jelas dengan susunan sumbu yang dihasilkan; pikirkan variabel rona sebagai dimensi ketiga di sepanjang sumbu kedalaman, di mana tingkat yang berbeda diplot dengan warna yang berbeda.
- Objek FacetGrid mengambil dataframe sebagai input dan nama-nama variabel yang akan membentuk dimensi baris, kolom, atau hue dari grid. Variabel harus kategoris dan data pada setiap tingkat variabel akan digunakan untuk segi di sepanjang sumbu itu.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
```

## Metode 2: Menggunakan metode PairGrid()



- subplot grid untuk memplot hubungan berpasangan dalam kumpulan data.
- Kelas ini memetakan setiap variabel dalam kumpulan data ke kolom dan baris dalam grid beberapa sumbu. Fungsi plotting tingkat sumbu yang berbeda dapat digunakan untuk menggambar plot bivariat di segitiga atas dan bawah, dan distribusi marjinal dari setiap variabel dapat ditunjukkan pada diagonal.
- Ini juga dapat mewakili tingkat konvensionalisasi tambahan dengan parameter hue, yang memplot subset data yang berbeda dalam warna yang berbeda. Ini menggunakan warna untuk menyelesaikan elemen pada dimensi ketiga, tetapi hanya menggambar subset di atas satu sama lain dan tidak akan menyesuaikan parameter hue untuk visualisasi spesifik seperti yang dilakukan fungsi tingkat sumbu yang menerima hue.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("flights")

plot = sns.PairGrid(data)
plot.map(plt.plot)

plt.show()
```

## Plot Relasional
Plot relasional digunakan untuk memvisualisasikan hubungan statistik antara titik data. Visualisasi diperlukan karena memungkinkan manusia untuk melihat tren dan pola dalam data. Proses memahami bagaimana variabel-variabel dalam kumpulan data saling berhubungan dan hubungannya disebut sebagai analisis statistik. Lihat artikel di bawah ini untuk informasi terperinci.

relplot()

Fungsi ini memberi kita akses ke beberapa fungsi tingkat sumbu lain yang berbeda yang menunjukkan hubungan antara dua variabel dengan pemetaan subset semantik. Itu diplot menggunakan metode relplot().

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# creating the relplot
sns.relplot(x='sepal_width', y='species', data=data)

plt.show()
```

## Categorical Plot


Plot Kategoris digunakan di mana kita harus memvisualisasikan hubungan antara dua nilai numerik. Pendekatan yang lebih khusus dapat digunakan jika salah satu variabel utama adalah kategorikal yang berarti variabel-variabel tersebut mengambil sejumlah nilai yang mungkin tetap dan terbatas.

Salah satu contoh categorical adalah barplot

kemudian ada juga boxplot

### boxplot

Ini menunjukkan distribusi data kuantitatif yang mewakili perbandingan antar variabel. boxplot menunjukkan kuartil dari dataset sementara whiskers meluas untuk menunjukkan sisa distribusi yaitu titik-titik yang menunjukkan adanya outlier. Itu dibuat menggunakan metode boxplot().

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

###violin plot
Ini mirip dengan boxplot kecuali bahwa ia memberikan visualisasi yang lebih tinggi dan lebih maju dan menggunakan estimasi kepadatan kernel untuk memberikan deskripsi yang lebih baik tentang distribusi data. Itu dibuat menggunakan metode violinplot().

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

###Strip plot


Ini pada dasarnya membuat plot pencar berdasarkan kategori. Itu dibuat menggunakan metode stripplot().

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.stripplot(x='species', y='sepal_width', data=data)
plt.show()
```

## Plot Distribution

Plot Distribusi digunakan untuk menguji distribusi univariat dan bivariat yang berarti distribusi yang melibatkan satu variabel atau dua variabel diskrit, salah satu contohnya adalah histogram dan lain-lain seperti :

###Distplot


Distplot pada dasarnya digunakan untuk kumpulan pengamatan univarian dan memvisualisasikannya melalui histogram yaitu hanya satu pengamatan dan karenanya kami memilih satu kolom tertentu dari kumpulan data.

###pairplot


Pairplot mewakili hubungan berpasangan di seluruh kerangka data dan mendukung argumen tambahan yang disebut hue untuk pemisahan kategoris. ### Regression Plot

Plot regresi terutama dimaksudkan untuk menambahkan panduan visual yang membantu untuk menekankan pola dalam kumpulan data selama analisis data eksplorasi. Plot regresi seperti namanya menciptakan garis regresi antara dua parameter dan membantu memvisualisasikan hubungan liniernya.

implot

Metode lmplot() dapat dipahami sebagai fungsi yang pada dasarnya membuat plot model linier. Ini menciptakan plot pencar dengan kecocokan linier di atasnya.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")

sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
```

Matrixplot

Plot matriks berarti memplot data matriks di mana diagram kode warna menunjukkan data baris, data kolom, dan nilai. Itu dapat ditampilkan menggunakan peta panas dan peta klaster.

```python
# Mengimpor paket
import seaborn as sns
import matplotlib.pyplot as plt

# Memuat dataset
data = sns.load_dataset("tips")

# Menghapus kolom non-numerik
data_numeric = data.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antara parameter yang berbeda
tc = data_numeric.corr()

sns.heatmap(tc)
plt.show()
```

```python
# Mengimpor paket
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Memuat dataset
data = sns.load_dataset("tips")

# Mengubah variabel kategorikal menjadi numerik menggunakan one-hot encoding
data_encoded = pd.get_dummies(data, drop_first=True)

# Menghitung korelasi antara parameter yang berbeda
tc = data_encoded.corr()

sns.heatmap(tc)
plt.show()
```
