```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
flights_data = sns.load_dataset("flights")
```

```python
flights_data.head()
```

## Transpose

```python
flights_data.T.head()
```

## Cross Tab

```python
percentage = pd.crosstab(index = flights_data["year"],
            columns = flights_data["month"],
            values = flights_data["passengers"],
            aggfunc = "sum",
            margins = True,
            margins_name = "Total",
            normalize = "index"
            )
```

```python
percentage.sum(axis = 1)
```

## Pivoting

```python
flights_data.pivot_table(index = "year",
                   columns = "month",
                   values = "passengers",
                   aggfunc = "max")
```

## Melting

```python
percentage
```

```python
flight_pct = percentage.reset_index().melt(id_vars = ["year"],
                value_vars = ["Jan", "Feb", "Mar"],
                var_name = "year",
                value_name = "Persentase Tahunan")
```

```python
hasil = percentage.reset_index().melt(id_vars = ["year"])
```

```python
flights_data.merge(hasil, on = ["year", "month"], how = "left")
```
